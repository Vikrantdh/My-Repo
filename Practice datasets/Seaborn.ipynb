```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill']) 
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
h = sns.FacetGrid(data=tips,col='time',row='smoker')
h.map(sns.distplot,'total_bill')
```

```python
h = sns.FacetGrid(data=tips,col='time',row='smoker')
h.map(plt.scatter,'total_bill','tip')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',aspect=0.6,size=8)
```
